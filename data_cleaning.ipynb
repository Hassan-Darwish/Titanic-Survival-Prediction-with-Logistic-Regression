```python
import numpy as np
import pandas as pd
```

```python
training_set = pd.read_csv('train.csv')
testing_set = pd.read_csv('test.csv')
```

```python
training_set.head()
```

```python
training_set['FamilySize'] = training_set['SibSp'] + training_set['Parch'] + 1
testing_set['FamilySize'] = testing_set['SibSp'] + testing_set['Parch'] + 1
```

```python
training_set.head()
```

```python
training_set.isna().sum()
```

```python
training_set['Age'].fillna(training_set['Age'].median(), inplace=True)
testing_set['Age'].fillna(testing_set['Age'].median(),inplace=True)
```

```python
testing_set[testing_set['Fare'].isna()]
```

```python
testing_set['Fare'].fillna(testing_set[testing_set['Pclass'] == 3]['Fare'].median(), inplace=True)
```

```python
testing_set.isna().sum()
```

```python
training_set['Embarked'].fillna(training_set['Embarked'].mode()[0], inplace=True)
```

```python
training_set.isna().sum()
```

```python
for series in (training_set,testing_set):
    series.drop(['PassengerId' ,'Name', 'Ticket', 'SibSp', 'Parch', 'Cabin'], axis=1,inplace=True)
```

```python
training_set = pd.get_dummies(training_set,columns=['Embarked'], dtype=int)
testing_set = pd.get_dummies(testing_set,columns=['Embarked'], dtype=int)
```

```python
for series in (training_set, testing_set):
    series['Sex'] = series['Sex'].map({"female" : 0, "male" : 1})
```

```python
training_set.head()
```

```python
for series in (training_set, testing_set):
    series['Pclass'] = (series['Pclass'] - series['Pclass'].min()) / (series['Pclass'].max() - series['Pclass'].min())
    series['Age'] = (series['Age'] - series['Age'].min()) / (series['Age'].max() - series['Age'].min())
    series['Fare'] = (series['Fare'] - series['Fare'].min()) / (series['Fare'].max() - series['Fare'].min())
    series['FamilySize'] = (series['FamilySize'] - series['FamilySize'].min()) / (series['FamilySize'].max() - series['FamilySize'].min())
```

```python
training_set.head()
```

```python
training_set.to_csv('train_cleaned.csv', index=False)
testing_set.to_csv('test_cleaned.csv', index=False)
```
